```python
import pandas as pd
import numpy as np
```

```python
books_wrangling = pd.read_csv('../data/goodreads_books.csv')
books_wrangling
```

```python
def drop_error(my_df, column_name, value):
    my_df = my_df[my_df[column_name] != value]
    return my_df
```

```python
books_wrangling = drop_error(books_wrangling, 'Title', 'error')
books_wrangling = drop_error(books_wrangling, 'Author', 'error')
books_wrangling = drop_error(books_wrangling, 'Rating', 'error')
books_wrangling = drop_error(books_wrangling, 'Ratings count', 'error')
books_wrangling = drop_error(books_wrangling, 'Reviews count', 'error')
books_wrangling = drop_error(books_wrangling, 'Pages', 'error')
books_wrangling = drop_error(books_wrangling, 'First published', 'error')
books_wrangling = drop_error(books_wrangling, 'Cover image', 'error')
books_wrangling = drop_error(books_wrangling, 'Genre', 'error')
books_wrangling = drop_error(books_wrangling, 'Book url', 'error')
books_wrangling
```

```python
def wrangling_split(df, columns, separator, element):
    for col in columns:
        df[col] = df[col].str.split(separator, expand = True)[element]
    return df
```

```python
book_filter_rating_pages_df = wrangling_split(books_wrangling, ['Ratings count', 'Reviews count', 'Pages'], " ", 0)
```

```python
book_filter_rating_pages_df = drop_error(book_filter_rating_pages_df, 'Pages', 'Unknown')
book_filter_rating_pages_df = drop_error(book_filter_rating_pages_df, 'Pages', 'Paperback')
book_filter_rating_pages_df = drop_error(book_filter_rating_pages_df, 'Pages', 'Hardcover')
book_filter_rating_pages_df = drop_error(book_filter_rating_pages_df, 'Pages', 'Audible')
book_filter_rating_pages_df = drop_error(book_filter_rating_pages_df, 'Pages', 'Audiobook')
book_filter_rating_pages_df = drop_error(book_filter_rating_pages_df, 'Pages', 'ebook')
book_filter_rating_pages_df = drop_error(book_filter_rating_pages_df, 'Pages', 'Audio')
book_filter_rating_pages_df = drop_error(book_filter_rating_pages_df, 'Pages', 'Kindle')
```

```python
book_filter_year_df = wrangling_split(book_filter_rating_pages_df, ['First published'], ",", 1)
```

```python
def wrangling_replace(df_replace, columns_replace, str_to_delete, str_to_insert):
    for col_replace in columns_replace:
        df_replace[col_replace] = df_replace[col_replace].str.replace(str_to_delete, str_to_insert)
    return df_replace
```

```python
def wrangling_to_type(df_to_type, columns_to_type, type_to_type):
    for col_to_type in columns_to_type:
        df_to_type[columns_to_type] = df_to_type[columns_to_type].astype(type_to_type)
    return df_to_type
```

```python
book_filter_to_int_df = wrangling_to_type(genre_filter_replace_df, ['Ratings count', 'Reviews count', 
                                                                  'Pages', 'First published'], int)
book_filter_to_int_df = wrangling_to_type(genre_filter_replace_df, ['Rating'], float)
book_filter_to_int_df
```
