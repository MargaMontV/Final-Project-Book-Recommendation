```python
import pandas as pd
import numpy as np
```

```python
books_wrangling = pd.read_csv('../data/goodreads_books.csv')
books_wrangling.head()
```

```python
def wrangling_split(df, columns, separator, element):
    for col in columns:
        df[col] = df[col].str.split(separator, expand = True)[element]
    return df
```

```python
book_filter_rating_pages_df = wrangling_split(books_wrangling, ['Ratings count', 'Reviews count', 'Pages'], " ", 0)
```

```python
book_filter_year_df = wrangling_split(book_filter_rating_pages_df , ['First published'], ",", 1)
```

```python
def wrangling_replace(df_replace, columns_replace, str_to_delete, str_to_insert):
    for col_replace in columns_replace:
        df_replace[col_replace] = df_replace[col_replace].str.replace(str_to_delete, str_to_insert)
    return df_replace
```

```python
book_filter_replace_df = wrangling_replace(book_filter_year_df, ['Ratings count', 'Reviews count'], 
                                           ",", "")
```

```python
def wrangling_to_int(df_to_int, columns_to_int):
    for col_to_int in columns_to_int:
        df_to_int[columns_to_int] = df_to_int[columns_to_int].astype(int)
    return df_to_int
```

```python
book_filter_to_int_df = wrangling_to_int(book_filter_replace_df, ['Ratings count', 'Reviews count', 
                                                                  'Pages', 'First published'])
book_filter_to_int_df
```

```python
book_filter_to_int_df.to_csv('../data/books_clean.csv', index=False)
```
